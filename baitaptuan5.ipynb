```python
#W5A3
def prime(n):
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return n > 1
```

```python
#W5A4
def perfect(n):
    cnt = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            cnt += i
            cnt += (n // i)
    if cnt == n:
        return True
    return False
```

```python
#W5A5
def search(arr, k):
    res = -1
    for i in range(len(arr)):
        if arr[i] == k:
            return i + 1
    return res
```

```python
#W5A6
def gt(n):
    res= 1
    while n != 0:
        res *= n
        n -= 1
    return res
```

```python
#W5A7
def maytinh(s):
    t = eval(s)
    return f'{t:.2f}'

print(maytinh(input()))
```

```python
#W5A8
def hamming(a, b):
    cnt = 0
    while ( a != 0) or ( b != 0 ):
        if (a%2) != (b % 2):
            cnt += 1
        a //= 2
        b //= 2
    return cnt

```

```python
#w5A9
def tong(n):
    res = 0
    while n != 0:
        res += (n % 10)
        n //= 10
    return res

print(tong(int(input())))
```
