```python
#W11A1
def reverse(path: str) -> int:
    file = open(path, "r")
    ans = file.readline().strip()[::-1]
    file.close()
    return ans
print(reverse("test.txt"))
```

```python
#11A2
def canWinNim(path: str) -> bool:
    file = open(path, "r")
    ans = int(file.readline().strip()) % 4 != 0
    file.close()
    return ans
print(canWinNim("test.txt"))
```

```python
#W11A3
def housesOfHogwarts(path: str):
    file = open(path, "r")
    n = int(file.readline().strip())
    for _ in range(n):
        print(file.readline().strip())
    file.close()
housesOfHogwarts("test.txt")

```

```python
#W11A4
def anagram(path : str):
    ls1, ls2 = [], []
    with open(path, "r") as f:
        line = f.readlines()
        for i in line[0].split():
            ls1.append(i.lower())
            
        for i in line[1].split():
            ls2.append(i.lower())
        
    if ls1.sort() == ls2.sort():
        print("TRUE")
    else:
        print("FALSE")
         
path = input()
anagram(path)
```

```python
#W11A5
def dem(path : str, character : str, k : str):
    cnt = 0
    with open(path, "w") as f:
        f.write(k)
    with open(path, "r") as f:
        t = f.read()
        cnt = t.count(character)
    return cnt


path = input()
character = input()
k = input()
print(dem(path, character, k))
```

```python
#W11A6
def cunt(path: str, character: str) -> int:
    file = open(path, "r")
    ans = sum(ch == character for ch in file.readline().strip())
    file.close()
    return ans
print(cunt("test.txt", "r"))
```

```python
#W11A8
def xau_fibo(path: str):
    file = open(path, "r")
    a, b = 'a', 'b'
    for _ in range(int(file.readline().strip()) + 1):
        print(a)
        a, b = b, a + b
    file.close()
xau_fibo("test.txt")
```

```python
#W11A9
import csv
def grade10(path: str):
    file = open(path, "r")
    for row in csv.reader(file):
        if row[0] != 'STT':
            print(row[1], row[2])
    file.close()
grade10("test.csv")
```

```python
#W11A10
def getMoney(path: int) -> list[int]:
    file = open(path, "r")
    ans = list(map(int, file.readline().strip().split()))
    file.close()
    return ans

def rob(houses: list[int]) -> int:
    dp1, dp2 = 0, 0
    for i in range(len(houses)):
        cur = max(dp1 + houses[i], dp2)
        dp1, dp2 = dp2, cur
    return dp2

print(rob(getMoney("test.txt")))
```

```python
#W11A11
def buoi(path: str):
    try:
        file = open(path, "r")
        lst = list(map(int, file.readline().split()))
        print(max(lst), min(lst))
        file.close()
    except:
        print("ngu")
buoi(input())
```

```python
#W11A12
def maximumProduct(path: str) -> int:
    file = open(path, "r")
    lst = sorted(list(map(int, file.readline().split())))
    file.close()
    return max(lst[0] * lst[1] * lst[-1], lst[-1] * lst[-2] * lst[-3])
print(maximumProduct("test.txt"))
```
