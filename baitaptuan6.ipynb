```python
#W6A3
def xoay_trai(k, ls):
    lis = []
    tmp = ls[:]  #shallow copy
    for i in range(k): #so lan thuc hien xoay trai
        if k < len(tmp): 
            lis.append(ls[0])
            ls.remove(ls[0])
        elif k == len(tmp):
            lis = []
        else:
            return xoay_trai(k - len(tmp), ls)
    ls.extend(lis)
    return tuple(ls)
ls = tuple(map(int, input().split()))
k = int(input())
ls = list(ls)
print(xoay_trai(k ,ls))
```

```python
#W6A8 : Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện của từng chuỗi và lưu vào dictionary
ls = list(input().split())
lis = []
for i in ls:
    if i not in lis:
        lis.append(i)
dic = {}
for j in lis:
    a = ls.count(j)
    dic[j] = a
print(dic)
```

```python
#W6A9
n1 = int(input())     #so phan tu cua dictionary thu nhat
dic1 = {}
for i in range(n1):
    key = input()
    value = int(input())
    dic1[key] = value
n2 = int(input())     #so phan tu cua dictionary thu hai
dic2 = {}
for i in range(n2):
    key = input()
    value = int(input())
    dic2[key] = value

res = dic1.copy()

for key in dic2:
    if key in res:
        res[key] += dic2[key]
    else:
        res[key] = dic2[key]


rks = sorted(res.keys())
sorted_res = {key : res[key] for key in rks}
print(sorted_res)
```
