```python
#W8A1:
class Cylinder:
    p = 3.14
    def __init__(self, r, h):
        self.r = r
        self.h = h

    def suface_area(self):
        return f"{(self.p * self.r * 2 * self.h):.2f}" 

    def vol(self):
        return f"{(self.p * self.r * self.r * self.h) :.2f}"

a, b = map(float, input().split())
hinh_tru = Cylinder(a, b)
print(hinh_tru.suface_area())
print(hinh_tru.vol())       
```

```python
#W8A3
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def goc(self):
        if self.x == self.y == 0: 
            return True
        return False
        
    def tung(self):
        if self.x == 0 and self.y != 0:
            return True
        return False
    
    def hoanh(self):
        if self.x != 0 and self.y == 0:
            return True
        return False
    
a, b = map(float, input().split())
M = Point(a, b)
print(M.goc())
print(M.tung())
print(M.hoanh())        
```

```python
#W8A4
class Caculator:
    def __init__(self, para_value1, para_value2):
        self.x = para_value1
        self.y = para_value2

    def add(self):
        return self.x + self.y

    def subtract(self):
        return self.x - self.y
    
    def product(self):
        return self.x * self.y

    def divide(self):
        if self.y == 0:
            return "khong chia duoc"
        else:
            return self.x / self.y
    
    def power(self):
        return self.x ** self.y
    
    def mod(self):
        return self.x % self.y
    
    def set_numbers(self, newx , newy ):
        self.x = newx
        self.y = newy

a, b = map(int, input().strip().split())
user = Caculator(a, b)
while True:
    print("                          1. Phep cong")
    print("                          2.Phep tru")
    print("                          3.Phep nhan")
    print("                          4.Phep chia")
    print("                          5.So mu")
    print("                          6.Phep mod")
    print("                          7.Cap nhat ")
    choice = int(input("Hay nhap chuc nang : "))
    if choice == 1:
        print(user.add())
    elif choice == 2:
        print(user.subtract())
    elif choice == 3:
        print(user.product())
    elif choice == 4:
        print(user.divide())
    elif choice == 5:
        print(user.power())
    elif choice == 6:
        print(user.mod())
    elif choice == 7:
        x, y = map(float, input().split())
        user.set_numbers(x, y)
    else:
        print("Ban co muon thoat chuong trinh khong :")
        a = input().strip().lower()
        if a == "yes":
            print("Ban da thoat khoi chuong trinh!")
            break

        
```

```python
#W8A5
class ShoppinhCart:
    def __init__(self):
        #feature1
        self.data = dict({})

    def feature2(self):
        key = input("Nhap ten san pham : ").strip()
        val = float(input("Nhap vao gia tien san pham : "))
        self.data[key] = val

    def feature3(self):
        print(f"So san pham trong gio hang truoc khi xoa : {len(self.data)}")
        key = input("Nhap vao ten san pham:").strip()
        self.data.pop(key)
        print(f'So san pham con lai trong gio hang la : {len(self.data)}')

    def feature4(self):
        if len(self.data) == 0:
            print("Gio hang cau ban rong!")
        else:
            print("Gio hang cua ban khong rong!")

    def feature5(self):
        bill = 0
        print("Tong so san pham trong gio hang la", len(self.data))
        for i in self.data.values():
            bill += i

        val = ""
        cnt, tmp = 0, 0

        for i in str(int(bill))[::-1]:
            val += i
            cnt += 1
            tmp += 1
            if cnt == 3 and tmp < len(str(int(bill))):
                val += '.'
                cnt = 0
        val = val[::-1]

        du = round((bill - int(bill)), 3)  

        lic = [i for i in str(du)]
        lic.pop(0)
        l = "".join(lic)
        val = val + l
        print(f'Tong so tien cac san pham trong gio hang la : {val} vnd')


    def feature6(self):
        if len(self.data) == 0:
            print("Gio hang cau ban rong!")
        else:
            print("Nhung san pham ban da them vao :")
            for i in self.data.keys():
                print(i, " : ", self.data[i], "vnd" )
        
    def feature7(self):
        self.data.clear()
        print("Bban da xoa toan bo san pham")

user = ShoppinhCart()
while True:
    
    print("                                         0. Thoat chuong trinh")
    print("                                         2. Thêm vào giỏ hàng một sản phẩm mới")
    print("                                         3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên để xóa)")
    print("                                         4. Kiểm tra xem giỏ hàng có rỗng không")
    print("                                         5. Tính tổng tiền giỏ hàng")
    print("                                         6. Hiển thị toàn bộ sản phẩm trong giỏ")
    print("                                         7. Xóa toàn bộ giỏ hàng")

    choice=  int(input())
    if choice == 2:
        user.feature2()
    elif choice == 3:
        user.feature3()
    elif choice == 4:
        user.feature4()
    elif choice == 5:
        user.feature5()
    elif choice == 6:
        user.feature6()
    elif choice == 7:
        user.feature7()
    else:
        print("Ban co muon thoat chuong trinh khong :")
        a = input().strip().lower()
        if a == "yes":
            print("Ban da thoat khoi chuong trinh!")
            break
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h, k):
        self.w = w
        self.h = h
        self.k = k
    
    def area(self):
        return f'{(self.h * self.w * self.k * self.k):.2f}'
    
    def perimeter(self):
        return f'{((self.h + self.w) * 2 * self.k):.2f}'

w, h, k = map(float, input().strip().split())
retangle1 = Rectangle(w, h, k)
print(f'{retangle1.area()} {retangle1.perimeter()}')
    
```
