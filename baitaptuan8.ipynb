```python
#W8A1:
class Cylinder:
    p = 3.14
    def __init__(self, r, h):
        self.r = r
        self.h = h

    def suface_area(self):
        return f"{(self.p * self.r * 2 * self.h):.2f}" 

    def vol(self):
        return f"{(self.p * self.r * self.r * self.h) :.2f}"

a, b = map(float, input().split())
hinh_tru = Cylinder(a, b)
print(hinh_tru.suface_area())
print(hinh_tru.vol())       
```

```python
#W8A3
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def goc(self):
        if self.x == self.y == 0: 
            return True
        return False
        
    def tung(self):
        if self.x == 0 and self.y != 0:
            return True
        return False
    
    def hoanh(self):
        if self.x != 0 and self.y == 0:
            return True
        return False
    
a, b = map(float, input().split())
M = Point(a, b)
print(M.goc())
print(M.tung())
print(M.hoanh())        
```

```python
#W8A4
class Caculator:
    def __init__(self, para_value1, para_value2):
        self.x = para_value1
        self.y = para_value2

    def add(self):
        return self.x + self.y

    def subtract(self):
        return self.x - self.y
    
    def product(self):
        return self.x * self.y

    def divide(self):
        if self.y == 0:
            return "khong chia duoc"
        else:
            return self.x / self.y
    
    def power(self):
        return self.x ** self.y
    
    def mod(self):
        return self.x % self.y
    
    def set_numbers(self, newx , newy ):
        self.x = newx
        self.y = newy

a, b = map(int, input().strip().split())
user = Caculator(a, b)
while True:
    print("                          1. Phep cong")
    print("                          2.Phep tru")
    print("                          3.Phep nhan")
    print("                          4.Phep chia")
    print("                          5.So mu")
    print("                          6.Phep mod")
    print("                          7.Cap nhat ")
    choice = int(input("Hay nhap chuc nang : "))
    if choice == 1:
        print(user.add())
    elif choice == 2:
        print(user.subtract())
    elif choice == 3:
        print(user.product())
    elif choice == 4:
        print(user.divide())
    elif choice == 5:
        print(user.power())
    elif choice == 6:
        print(user.mod())
    elif choice == 7:
        x, y = map(float, input().split())
        user.set_numbers(x, y)
    else:
        print("Ban co muon thoat chuong trinh khong :")
        a = input().strip().lower()
        if a == "yes":
            print("Ban da thoat khoi chuong trinh!")
            break

        
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h, k):
        self.w = w
        self.h = h
        self.k = k
    
    def area(self):
        return f'{(self.h * self.w * self.k * self.k):.2f}'
    
    def perimeter(self):
        return f'{((self.h + self.w) * 2 * self.k):.2f}'

w, h, k = map(float, input().strip().split())
retangle1 = Rectangle(w, h, k)
print(f'{retangle1.area()} {retangle1.perimeter()}')
    
```

```python
#W8A7
import math
class Fraction:
    def __init__(self, tu, mau):
        self.tu = tu
        self.mau = mau

    def __str__(self):
        return f'{self.tu}/{self.mau}'

    def Rutgon(self):
        g = math.gcd(self.tu, self.mau)
        self.tu //= g
        self.mau //= g
        

    def __add__(self, other):
        return Fraction(self.tu * other.mau + self.mau * other.tu , (self.mau * other.mau))
    
    def __sub__(self, other):
        return Fraction(self.tu * other.mau - self.mau * other.tu , (self.mau * other.mau))
    
    def __mul__(self, other):
        return Fraction((self.tu * other.tu) , (self.mau * other.mau))
    
    def __truediv__(self, other):
        return Fraction((self.tu * other.mau) , (self.mau * other.tu))
    

if __name__ == "__main__":
    a = list(input().split())
    p1 = Fraction(int(a[0]), int(a[1]))
    p2 = Fraction(int(a[3]), int(a[4]))
    if a[2] == "+":
        p3 = p1 + p2
        p3.Rutgon()
        print(p3)
    elif a[2] == "-":
        p3 = p1 - p2
        p3.Rutgon()
        print(p3)
    elif a[2] == "*":
        p3 = p1*p2
        p3.Rutgon()
        print(p3)
    else:
        p3 = p1 / p2
        p3.Rutgon()
        print(p3)
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, x):
        self.balance = self.balance + x
    
    def withdraw(self, y):
        self.balance = self.balance - y
    
    def __str__(self):
        return f"{self.balance:.2f}"
    
if __name__ == "__main__":
    t = input().split()
    acc = BankAccount(t[0], float(t[1]))
    n = int(input())
    for i in range(n):
        a = input().split()
        if a[0] ==  "WITHDRAW":
            acc.withdraw(float(a[1]))
        else:
            acc.deposit(float(a[1]))
    print(acc)
    
```

```python
#W8A12
class SoPhuc:
    def __init__(self, thuc, ao):
        self.thuc = thuc
        self.ao = ao

    def __str__(self):
        return f'{self.thuc}  {self.ao}'

    def __mul__(self, other):
        thuc = self.thuc * other.thuc - self.ao * other.ao
        ao = self.thuc * other.ao + self.ao * other.thuc
        return SoPhuc(thuc, ao)             #khong gan nguoc lai cho self
    
if __name__ == "__main__":
    a = list(map(int, input().split()))
    b = list(map(int, input().split()))
    p1 = SoPhuc(a[0], a[1])
    p2 = SoPhuc(b[0], b[1])
    p3 = p1 * p2
    print(p3)
```

```python
#W8A13
import math
class Triangle:
    def __init__(self, egde1, egde2, egde3):
        self.egde1 = egde1
        self.egde2 = egde2
        self.egde3 = egde3
    def getArea(self):
        if self.egde1 <= 0 or self.egde2 <= 0 or self.egde3 <= 0 or self.egde1 + self.egde2 < self.egde3 or self.egde1 + self.egde3 < self.egde2 or self.egde3 + self.egde2 < self.egde1:
            return "INVALID"
        else:
            p = (self.egde1 + self.egde2 + self.egde3)/2 
            return f'{(math.sqrt(p*(p - self.egde1)*(p-self.egde2)*(p-self.egde3))):.2f}'

if __name__ == "__main__":
    a = list(map(int, input().split()))
    tri = Triangle(a[0], a[1], a[2])
    print(tri.getArea())
```
