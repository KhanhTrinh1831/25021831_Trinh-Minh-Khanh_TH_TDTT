```python
'''
try: Chứa đoạn code có khả năng gây ra lỗi.

except: Chứa code xử lý nếu lỗi xảy ra trong khối try.

else: Chứa code chỉ chạy khi KHÔNG có lỗi nào xảy ra trong khối try.

finally: Chứa code LUÔN LUÔN chạy, bất kể có lỗi hay không (thường dùng để dọn dẹp tài nguyên)
'''
```

```python
#W11A1
a, b = map(int, input().split())
try:
    print(f'{(a/b):.2f}')
except:
    print("Loi chia khong")
```

```python
#W11A2
try:
    ls = list(map(int, input().split()))
    n = int(input())
    print(ls[n])
except:
    print("Truy cap chi so sai")
```

```python
#W11A3
try:    
    def tong_chu_so(n):
        cnt = 0
        while n != 0:
            cnt += (n % 10) **2
            n //= 10
        return cnt

    def happy_number(n):
        if tong_chu_so(n) == 1:
            return("YES")
        return happy_number(tong_chu_so(n))
    n =  int(input())
    print(happy_number(n))
except:
    print("NO")
```

```python
#W11A4

try :                                   #khoi code thuc hien
    m, n = map(int, input().split())
    res = m + n
except ValueError:                 
    print("Invalid Input")
else:                                  #code chay khi khong co loi nao xay ra trong try
    print(res)
finally:                               #luon xay ra bat ke co loi hay khong
    print("Ket thuc chuong trinh")
```

```python
#W11A10
try:
    ls = list(map(int, input().strip().split()))
    if not ls:
        raise ValueError
    n = int(ls[0])
    lis = ls[1:]
    if len(lis) != n:
        raise ValueError
    
    if n != len(list(set(lis))):
        raise ValueError
    
    for i in lis:
        if i <= 0:
            raise ValueError
        
    print(sum(lis))

except ValueError:
    print("Mang khong hop le")
    
```

```python
#W11A11
import re
try:
    n = input().strip()
    if n:
        k = int(n)
        for _ in range(k):
            t = input()
            try:
                re.compile(t)
                print("True")
            except re.error:
                print("False")
except:
    pass
```
