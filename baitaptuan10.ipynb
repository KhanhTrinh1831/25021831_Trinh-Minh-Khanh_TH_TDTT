```python
#W9A1
def merge(l, r): 
    i, j = 0, 0
    res = []
    while i < len(l) and j < len(r):
        if l[i] <= r[j]:
            res.append(l[i])
            i += 1
        else:
            res.append(r[j])
            j += 1

    res.extend(l[i:])
    res.extend(r[j:])
    return res

l = list(map(int, input().split()))
r = list(map(int, input().split()))
print(*merge(l, r))
```
